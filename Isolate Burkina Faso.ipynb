```python
burkina_df = burkina_df.drop(['EVENT_ID_CNTY','TIMESTAMP','ISO', 'REGION', 'COUNTRY','ADMIN1', 'ADMIN2', 'ADMIN3', 'GEO_PRECISION'], axis=1)
burkina_df.head()
```

```python
burkina_df = burkina_df.drop(['TIME_PRECISION','TAGS'], axis=1)
```

```python
burkina_df.columns
```

```python
burkina_df.head()
```

```python
burkina_df.dtypes
```

```python
burkina_df['EVENT_DATE'] = pd.to_datetime(burkina_df['EVENT_DATE'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# group the dataframe by EVENT_DATE and count the number of events in each group
grouped_df = burkina_df.groupby('EVENT_DATE').size().reset_index(name='EVENT_COUNT')

# create a new figure
plt.figure(figsize=(20, 6))

# create a line plot of the total number of events over time
plt.plot(grouped_df['EVENT_DATE'], grouped_df['EVENT_COUNT'], linewidth=1, color='red', alpha=0.8)

# add title and labels
plt.title('Total Events over Time')
plt.xlabel('Event Date')
plt.ylabel('Number of Events')

# show the plot
plt.show()
```

```python
# create a new figure
plt.figure(figsize=(20, 6))

# group the events by their date and compute the count of events for each date
grouped_df = burkina_df.groupby('EVENT_DATE').count()['EVENT_TYPE']

# filter the data to show events between the first election of Kabore and the latest info
kabore_first_election = '2015-11-29'
latest_info = grouped_df.index.max()
filtered_df = grouped_df.loc[kabore_first_election:latest_info]

# create a line plot of the filtered data
plt.plot(filtered_df.index, filtered_df)

# add title and labels
plt.title('Events in Burkina Faso between Kabore\'s first election and the latest information')
plt.xlabel('Date')
plt.ylabel('Number of Events')

# show plot
plt.show()
```

```python
# group the events by their date and compute the count of events for each date
grouped_df = burkina_df.groupby('EVENT_DATE').count()['EVENT_TYPE']

# filter the data to show events between the first election of Kabore and the latest info
kabore_zerbo = '2021-11-29'
latest_info = grouped_df.index.max()
filtered_df = grouped_df.loc[kabore_zerbo:latest_info]

# create a line plot of the filtered data
sns.lineplot(x=filtered_df.index, y=filtered_df)

# add title and labels
plt.title('Events in Burkina Faso between Kabore\'s Last Government and Now')
plt.xlabel('Event Date')
plt.ylabel('Event Count')

# add markers at specific dates
plt.axvline(pd.to_datetime('2022-01-24'), color='red', linestyle='dashed')
plt.axvline(pd.to_datetime('2022-09-30'), color='blue', linestyle='dashed')

# show the plot
plt.show()
```
